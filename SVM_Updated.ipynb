# Linear Kernel with MAD Filtering

Correlation filtering

```python
cor_matrix = df_unfiltered.corr().abs()
print(cor_matrix)
```

```python
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

```python
to_drop_corr = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(to_drop_corr)
```

```python
def mean_absolute_deviation(col):
    mean = col.mean()
    total = 0
    for i in col:
        total += abs(i - mean)
    return total/len(col)

lst = list(merged.columns)
lst.remove('0')
lst.remove('class')
```

```python
all_devs = []
for i in lst:
    all_devs.append((i, mean_absolute_deviation(merged[i])))
to_keep = [i for i in all_devs if i[1] > 0.75]
to_keep = [i[0] for i in to_keep]
```

```python
to_drop_corr = ['6_x', '7_x', '8_x', '13_x', '14_x', '15_x', '18_x', '20_x', '21_x', '22_x', '25_x', '27_x', '28_x', '33_x', '36_x', '850', '1180', '1377', '1403', '1591', '1941', '1952', '2390', '2436', '2448', '2740', '2816', '2825', '2888', '2945', '3146', '3286', '3465', '3506', '3733', '3873', '3889', '3900', '3901', '4476', '4600', '4780', '4864', '4965', '4968', '5020', '5165', '5219', '5397', '5399', '5526', '5748', '6396', '6570', '6806', '6851', '7279', '7781', '7953']
for i in to_drop_corr:
    if i in to_keep:
        to_keep.remove(i)
```

```python
filtered_merged = merged[['0'] + to_keep + ['class']]
```

```python
filtered_merged.to_csv('AfterMADFiltering.csv', index=False)
```

```python
df_features = merged[to_keep]
df_targets = merged['class']
```

```python
df_features
```

```python
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
scaler = preprocessing.StandardScaler().fit(df_features)
df_features_scaled = scaler.transform(df_features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features_scaled, np.ravel(df_targets),
                                                            test_size = 0.30, random_state = 344)
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['linear']}
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
print(classification_report(y_test, grid_predictions))
```

```python
print(confusion_matrix(y_test, grid_predictions))
```

```python
model_lin = SVC(C=0.1, gamma=1, kernel='linear')
scores_lin = cross_val_score(model_lin, df_features_scaled, np.ravel(df_targets), cv=5)
np.mean(scores_lin)
```

# PseudoDistance Calculation using model

```python
model_linear = SVC(C=0.1, gamma=1, kernel='linear')
model_linear.fit(df_features_scaled, np.ravel(df_targets))
```

```python
predictions = model_linear.predict(df_features_scaled)
```

```python
confusion_matrix(np.ravel(df_targets), predictions)
```

```python
y = model_linear.decision_function(df_features_scaled)
w_norm = np.linalg.norm(model_linear.coef_)
distance_from_hyperplane = y / w_norm
```

```python
sigmoid_distance = 1/(1 + np.exp(-distance_from_hyperplane))
```

```python
sigmoid_distance
```

```python
distance_data = pd.DataFrame()
distance_data['image'] = merged['0']
distance_data['dist'] = sigmoid_distance
```

```python
distance_data.to_csv('scan_distfromhyperplane.csv', index=False)
```
