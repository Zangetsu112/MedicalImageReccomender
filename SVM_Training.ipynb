```python
import os 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
data_ML = pd.read_csv('TrainingSet_MLFeatures.csv')
data_DL = pd.read_csv('TrainingSet_DLFeatures.csv')
```

```python
data_ML
```

```python
data_DL
```

```python
to_change = [i for i in data_ML.columns if data_ML[i].dtype != np.float64]
to_change.remove('0')
```

```python
for i in to_change:
    data_ML[i] = data_ML[i].map(lambda x: str(x)[1:-1]).astype(np.float64)
```

```python
merged = pd.merge(data_ML, data_DL, how='inner', on='0')

def convert_to_class(val):
    if val[:5] == "COVID":
        return 1
    else: return 0

merged['class'] = merged['0'].map(convert_to_class)
```

```python
merged
```

```python
def mean_absolute_deviation(col):
    mean = col.mean()
    total = 0
    for i in col:
        total += abs(i - mean)
    return total/len(col)

lst = list(merged.columns)
lst.remove('0')
lst.remove('class')
```

```python
all_devs = []
for i in lst:
    all_devs.append((i, mean_absolute_deviation(merged[i])))
to_keep = [i for i in all_devs if i[1] > 0.75]
to_keep = [i[0] for i in to_keep]
```

```python
filtered_merged = merged[['0'] + to_keep + ['class']]
```

```python
filtered_merged.to_csv('AfterMADFiltering.csv', index=False)
```

```python
df_features = merged[to_keep]
df_targets = merged['class']
```

```python
df_features
```

```python
df_targets
```

```python
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
scaler = preprocessing.StandardScaler().fit(df_features)
df_features_scaled = scaler.transform(df_features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features_scaled, np.ravel(df_targets),
                                                            test_size = 0.30, random_state = 344)
```

# RBF Kernel

```python
param_grid = {'C': [0.1, 1, 10, 100], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
print(classification_report(y_test, grid_predictions))
```

```python
print(confusion_matrix(y_test, grid_predictions))
```

# Linear Kernel

```python
param_grid2 = {'C': [0.1, 1, 10, 100], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['linear']}
grid2 = GridSearchCV(SVC(), param_grid2, refit = True, verbose = 3)
grid2.fit(X_train, y_train)
```

```python
print(grid2.best_params_)
print(grid2.best_estimator_)
```

```python
grid_predictions2 = grid2.predict(X_test)
print(classification_report(y_test, grid_predictions2))
```

```python
print(confusion_matrix(y_test, grid_predictions2))
```

# Polynomial Kernel

```python
param_grid3 = {'C': [0.1, 1, 10, 100], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['poly']}
grid3 = GridSearchCV(SVC(), param_grid3, refit = True, verbose = 3)
grid3.fit(X_train, y_train)
```

```python
print(grid3.best_params_)
print(grid3.best_estimator_)
```

```python
grid_predictions3 = grid3.predict(X_test)
print(classification_report(y_test, grid_predictions3))
```

```python
print(confusion_matrix(y_test, grid_predictions3))
```

# Testing different SVCs with cross fold

```python
model = SVC(C=100, gamma=0.0001)
scores = cross_val_score(model, X_train, y_train, cv=25)
np.mean(scores)
```

```python
model_lin = SVC(C=1, gamma=1, kernel='linear')
scores_lin = cross_val_score(model_lin, X_train, y_train, cv=25)
np.mean(scores_lin)
```

```python
model_poly = SVC(C=0.1, gamma=0.01, kernel='poly')
scores_poly = cross_val_score(model_poly, X_train, y_train, cv=25)
np.mean(scores_poly)
```

# Linear model training and testing

```python
model_linear = SVC(C=1, gamma=1, kernel='linear')
model_linear.fit(X_train, y_train)
```

```python
predictions = model_linear.predict(df_features_scaled)
```

```python
confusion_matrix(np.ravel(df_targets), predictions)
```

```python
y = model_linear.decision_function(df_features_scaled)
w_norm = np.linalg.norm(model_linear.coef_)
distance_from_hyperplane = y / w_norm
```

```python
sigmoid_distance = 1/(1 + np.exp(-distance_from_hyperplane))
```

```python
sigmoid_distance
```

```python
distance_data = pd.DataFrame()
distance_data['image'] = merged['0']
distance_data['dist'] = sigmoid_distance
```

```python
distance_data.to_csv('scan_distfromhyperplace.csv', index=False)
```
